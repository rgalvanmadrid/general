**¿Cómo escoger el tamaño del pixel cuando se lee el modelo?:**\
Para las imágenes que se usan como un modelo, siempre hay que tomar una decisión de a qué tamaño angular en el cielo corresponde cada pixel.\
El elemento de resolución de un telescopio es el llamado Point Spread Function (o PSF), al cual en radiointerferometría también se le llama beam o haz sintetizado. El PSF debe de estar muestreado por al menos 2 pixeles (muestreo de Nyquist). 
Entonces, ¿cómo escoger la escala angular del pixel? Asignandole aproximadamente 1/2 del tamaño del PSF más pequeño con el vayamos a jugar con el friendlyVRI. Por ejemplo, la máxima resolución angular de ALMA en la Banda 6 (230 GHz, 1.3 mm) es 20 milliarcsec = 0.02 arcsec. Para este caso, habría que poner que el tamaño del pixel es 0.01 arcsec. 

Las resoluciones angulares y otros parámetros básicos de ALMA y el VLA pueden encontrarse en:\
https://almascience.nrao.edu/about-alma/alma-basics \
https://science.nrao.edu/facilities/vla/docs/manuals/oss/performance/resolution
